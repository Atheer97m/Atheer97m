```python
# droping ID column 
data = data.drop('ID', axis=1) 
```

```python
#Assigning the featurs as X and trarget as y
X = data.drop('Reached.on.Time_Y.N', axis=1)
y = data['Reached.on.Time_Y.N']
```

```python
ordinal = [
    'Gender',
    'Product_importance'
]

nominal = [
    'Warehouse_block',
    'Mode_of_Shipment'
]

scal = [
    'Cost_of_the_Product',
    'Discount_offered',
    'Weight_in_gms'
]
```

```python
# Building pipelins
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler


transformer_for_ordinal = Pipeline([
    ('ordinal', OrdinalEncoder(categories='auto'))
])
transformer_for_nominal = Pipeline([
    ('nominal', OneHotEncoder())
])

transformer_for_scal = Pipeline([
    ('scaler', MinMaxScaler())
])
```

```python
Transformer = ColumnTransformer(transformers=[
    ('ordinal', transformer_for_ordinal, ordinal),
    ('nominal', transformer_for_nominal, nominal),
    ('scal', transformer_for_scal, scal)
], sparse_threshold=0)
```

```python
data_trans = pd.DataFrame(Transformer.fit_transform(X))
data_trans['y'] = y

data_trans
```

```python
# correlation matrix

plt.figure(figsize = (18, 9), facecolor='#fbe7dd')
sns.heatmap(data_trans.corr(), annot = True, vmin=-1, vmax=1, fmt = '.2f', annot_kws={'size' : 15}, cmap= 'viridis')
plt.show()
```

```python
corr = data_trans.corr()
```

```python
X_tran = data_trans[
    corr[abs(corr['y']) > 0.005]['y'].index
]

X_tran = X_tran.drop(['y'], axis=1)

X_tran
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_tran, y, train_size=0.75, shuffle=True, random_state=42)
```

```python
# logistic regression model
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error, r2_score
model = LogisticRegression()
model.fit(X_train, y_train)
predictions = model.predict(X_test)
print("Training Accuracy :", model.score(X_train, y_train))
print("Testing Accuracy :", model.score( X_test, y_test))
print("Mean squared error: %.2f" % mean_squared_error(y_test, predictions))
# The coefficient of determination: 
print("r2 score: %.2f" % r2_score(y_test, predictions))
```

```python
# linear regression model
from sklearn.linear_model import LinearRegression
la_model = LinearRegression()
la_model.fit(X_train, y_train)
la_predictions = la_model.predict(X_test)
print("Training Accuracy :", la_model.score(X_train, y_train))
print("Testing Accuracy :", la_model.score( X_test, y_test))
print("Mean squared error: %.2f" % mean_squared_error(y_test, predictions))
# The coefficient of determination: 
print("r2 score: %.2f" % r2_score(y_test, predictions))
```
